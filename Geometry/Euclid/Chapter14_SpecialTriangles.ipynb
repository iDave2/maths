# Chapter 14 - Special Triangles


## Definition - Congruent Triangles

> Two triangles are **congruent** if and only if they are similar triangles for which the ratio of similitude is $k=1$.  If $\triangle ABC$ is congruent to $\triangle DEF$, it is denoted by the expression $\triangle ABC \cong \triangle DEF$.

### Theorem 14-1

> Corresponding parts (i.e., angles and sides) of congruent triangles are congruent.  (This is often abbreviated C.P.C.T.C.)

Yes, I say that all the time.

### Theorem 14-2

> If each side of one triangle is congruent to the corresponding side of another triangle, then the triangles are congruent.  (Often abbreviated S.S.S?)

### Corollary 14-2.1

> If any two sides of a right triangle are congruent to the corresponding sides of another right triangle, the triangles are congruent.

### Theorem 14-3

> If two sides and the included angle of one triangle are congruent to the corresponding two sides and included angle of another triangle, then the triangles are congruent.  (Abbreviated S.A.S.)

### Theorem 14-4

> If two angles and a side of one triangle are congruent to the corresponding two angles and side of another triangle, then the triangles are congruent.  (Abbreviated S.A.A.)

### Corollary 14-4.1

> If two angles and the included side of one triangle are congruent to the corresponding angles and the included side of another triangle, then the triangles are congruent.  (Abbreviated A.S.A.)

This is redundant with Theorem 14-4, it adds nothing new except perhaps another swell acronym.

### Corollary 14-4.2

> If any side and an acute angle of one right triangle are congruent to the corresponding side and acute angle of another right triangle, then the triangles are congruent.

## Definition - Angle Bisector

> For any angle, an **angle bisector** is a ray interior to the angle such that its endpoint is the vertex of the angle and it determines congruent angles with the sides of the angle.

### Theorem 14-5

> Every angle has one and only one bisector.

### Theorem 14-6

> A point is on the angle bisector of an angle if and only if it is equidistant from the sides of the angle.  (Remember that the distance from a point to a line is the length of the segment fro the point perpendicular to the line.)

### Theorem 14-7

> The three angle bisectors of the angles of a triangle intersect at a point.

## Definition - Incenter

> The point of intersection of the three angle bisectors of the angles of a triangle is called the **incenter** of the triangle.

## Definition - Concurrent Lines

> A set of lines is **concurrent** if and only if the lines intersect in exactly one point.

### Theorem 14-8

> The lines including the altitudes of a triangle are concurrent.

## Definition - Orthocenter

> The point of intersection of the three lines containing the altitudes of a triangle is called the **orthocenter**.

### Theorem 14-9

> The three perpendicular bisectors of the sides of a triangle are concurrent.

## Definition - Circumcenter

> The point of intersection of the perpendicular *bisector* of the sides of a triangle is called the **circumcenter** of the triangle.  The circumcenter of a triangle is usually denoted by the letter $J$.

This term may be more commonly defined relative to the center of a circle that touch each point of a so-called *cyclic polygon*.  Not all polygons are cyclic.

## Definition - Median

> A median of a triangle is a segment whose endpoints are a vertex of the triangle and the midpoint of the side opposite that vertex.

### Theorem 14-10

> The medians of a triangle are concurrent.

Uhmm ... same as incenter?  No.  Angle bisectors are equidistant from sides; midpoints are equidistant from endpoints of a segment.  Rotate one side of an equilateral triangle about its midpoint while keeping opposite angle constant and adjusting segment lengths as needed.  The midpoint then moves away from the opposite angle's bisector.  Q.E.C.

## Definition - Centroid

> The point of intersection of the medians of any triangle is called the **centroid** or the **center of gravity**.

### Corollary 14-10.1

> The centroid of a triangle is two thirds of the distance from a vertex to the midpoint of the side opposite that vertex.

### Theorem 14-11

> If $P$ is between $P_1 = (x_1,y_1)$ and $P_2 = (x_2,y_2)$ and
>
> $$ {P_1 P\over PP_2} = r, $$
>
> then
>
> $$ P = \Bigl( {x_1 + rx_2\over 1+r}, {y_1 + ry_2\over 1+r} \Bigr). $$

### Theorem 14-12

> The median to the hypotenuse of any right triangle is equal in length to $\Frac1/2$ the length of the hypotenuse.

## Definition - Scalene & Isosceles Triangles

> A **scalene triangle** is a triangle with no two sides congruent.
>
> An **isosceles triangle** is a triangle with at least two congruent sides.  The angle formed by the two congruent sides of an isosceles triangle is **the vertex angle** of the triangle.  The remaining angles are the **base angles**.

### Theorem 14-13

> A triangle is isosceles if and only if its base angles are congruent.  (This is often referred to as the "Isosceles Triangle Theorem.")

## Definition - Equilateral Triangle

> An **equilateral triangle** is a triangle with three congruent sides.

### Theorem 14-14

> A triangle is equilateral if and only if it is equiangular.

### Theorem 14-15

> In an isosceles right triangle, the length of the side opposite the vertex angle (the hypotenuse) is $\sqrt 2$ times the length of a leg of the triangle.

### Theorem 14-16

> In a $30\Degrees - 60\Degrees - 90\Degrees$ triangle, the length of the side opposite the $30\Degrees$ angle is one-half the length of the hypotenuse.

### Theorem 14-17

> In a $30\Degrees - 60\Degrees - 90\Degrees$ triangle, the length of the side opposite the $60\Degrees$ angle is ${\sqrt 3}\,/\,2$ times the length of the hypotenuse.

The remaining theorems are "obvious," the proofs tricky, try it.

### Theorem 14-18

> If two sides of a triangle are not congruent, then the angles opposite these sides are not congruent and the larger angle is opposite the longer side.

### Theorem 14-19

> (Converse of Theorem 14-18.)  If two angles of a triangle are not congruent, then the sides opposite these angles are not congruent and the longer side is opposite the larger angle.

### Theorem 14-20

> The sum of the lengths of any two sides of a triangle is greater than the length of the third side.  (This is often called the triangle inequality theorem.)

### Theorem 14-21

> If two sides of one triangle are congruent respectively to two sides of a second triangle, and the included angle of the first triangle is larger than the included angle of the second, then the third side of the first triangle is larger than the third side of the second triangle.  (This is sometimes called the "Hinge Theorem.")

### Theorem 14-22

> (Converse of Theorem 14-21.)  If two sides of one triangle are congruent respectively to two sides of a second triangle, and the third side of the first triangle is longer than the third side of the second, then the included angle of the first triangle is larger than the included angle of the second triangle.
