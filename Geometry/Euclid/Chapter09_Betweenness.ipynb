# Chapter 9 - Betweenness

$\def\Between#1#2#3{#1\negthinspace-\negthinspace#2\negthinspace-\negthinspace#3}$
$\def\Line#1{\buildrel\longleftrightarrow\over{#1}}$
$\def\Segment#1{\overline{#1}}$
$\def\Ray#1{\kern-.1em\vec{\kern.1em #1}}$

## Definition - Between

> For three different points $A$, $B$, and $C$ on the same line, $B$ is **between** $A$ and $C$ if and only if $AB+BC=AC$.  (Recall that $AB$ is shorthand for a segment's length.)  The expression $\Between A B C$ is used to indicate $B$ is between $A$ and $C$.

### Theorem 9-1

> Between any two different points $P$ and $Q$, there is at least one other point.

### Theorem 9-2

> For any two different points $P$ and $Q$, there is a point $R$ such that $\Between P Q R$.

## Definition - Line Segment

> If $A$ and $B$ are different points, then the union of $A$ and $B$ with all the points between them is called a **line segment** or **segment** of the line $\Line{AB}$.  $A$ and $B$ are called the **endpoints** of the segment.  The segment is denoted by $\Segment{AB}$.

It is good to finally define a line since it has been used since Genesis.  This becomes rather confusing, a sense of introducing abstraction late for abstraction's sake?

## Definition - Segment Measure

> The **measure** of $\Segment{AB}$ is the distance from $A$ to $B$ and is written simply as $AB$.

## Definition - Ray

> If $A$ and $B$ are different points, then the union of $\Segment{AB}$ with all the points $P$ such that $\Between A B P$ is called a **ray**.  Point $A$ is the **endpoint** of the ray.  The ray is denoted by $\Ray{AB}$.

## Definition - Congruent Segments

> Segments $\Segment{AB}$ and $\Segment{CD}$ are **congruent** if and only if they have the same measure; that is, iff $AB = CD$.  Congruent segments are denoted by $\Segment{AB}\cong\Segment{CD}$.

Here is the magic work *measure* again.  So one can speak about the measure of a segment (its length) or the measure of an angle (its degrees of arc), etc.  Point here is that congruence differs from equality and refers, or will soon refer, to a sameness of shape.

### Theorem 9-2.5

> If two congruent segments have equal slopes or no slopes, then the difference of the abscissas and ordinates of their endpoints, when taken in the same order, are equal.

Imagine translating a segment in the plane without rotating it.  The endpoints remain in the same location relative to segment midpoint.

## Definition - Direction Numbers

> The real numbers $x_2-x_1$ and $y_2-y_1$ of Theorem 9-2.5 are called the **x direction number** and the **y direction number** respectively of the given line segment.

## Definition - Midpoint

> Point $P$ is a midpoint of $\Segment{AB}$ if and only if
>
> 1. $P$ lies between $A$ and $B$ ($\Between A P B$) and
> 2. $AP = PB$.

### Theorem 9-3

> If $P = (x,y)$ is the midpoint of $\Segment{AB}$ with $A = (a,b)$ and $B = (c,d)$, then
>
> $$ (x,y) = \Bigl( {a+c\over 2} , {b+d\over 2} \Bigr). $$

### Theorem 9-4

> (Converse of Theorem 9-3.)  If $A = (a,b)$, $B=(c,d)$, and $P$ are points such that
>
> $$ P = \Bigl( {a+c\over2}, {b+d\over2} \Bigr), $$
>
> then $P$ is the midpoint of $\Segment{AB}$.

## Definition - Bisect

> If point $P$ is the midpoint of a segment, then it is said to **bisect** the segment.

## Definition - Bisector

> Any line or segment that intersects the midpoint and is distinct from another given segment is called a **bisector** of the given segment.

## Definition - Perpendicular Bisector

> The **perpendicular bisector** of a given segment is a line or segment perpendicular to it that passes through its midpoint.

### Theorem 9-5

> If a point is on the perpendicular bisector of a segment, then it is equally distant (equidistant) from the endpoints of the segment.

### Theorem 9-6

> (Converse of Theorem 9-5.)  If points $A$, $B$, and $P$ are different points and $AP = PB$, then $P$ is on the perpendicular bisector of $\Segment{AB}$.
