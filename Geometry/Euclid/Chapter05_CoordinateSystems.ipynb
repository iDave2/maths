## Postulate 1 - Points & Real Numbers

> The points of a line can be placed in a correspondence with the real numbers in such a way that,
>
> 1. to every point of the line there corresponds exactly one real number;
> 2. to every real number there corresponds exactly one point of the line.

## Definition - One Dimensional Coordinates

> A correspondence of the sort described in Postulate 1 is called a **one dimensional coordinate system**.  The number corresponding to a given point is called the **coordinate** of the point.

## Postulate 2 - Restatement of Axiom 1

> For every two points there is exactly one line that contains both points.

This is Euclid's [Axiom #1](Chapter02_HistoricalGeometry.ipynb).

## Postulate 3 - Axis Translation

> Given two points $P$ and $Q$ of a line, the coordinate system can be chosen in such a way that the coordinate of $P$ is zero and the coordinate of $Q$ is positive.

In other words, coordinates may be adjusted to simplify problem solution and this is not considered cheating.

## Definition - Parallel Lines

> Two lines in a plane are **parallel** if and only if they have no points in common; that is, iff they do not intersect.

## Definition - Class of Lines

> A **class of lines** is the set of all lines in a plane parallel to a given line in the plane.

## Definition - Two Dimensional Coordinates

> Let $L_1$ and $L_2$ be a pair of intersecting lines and consider the class of lines parallel to each.  The resulting network of lines determine a **two dimensional coordinate system** or **grid system**.

## Postulate 4 - Axis Primer, Grids

> If class $H$ and class $V$ are two different classes of lines, then each line of class $H$ has exactly one point in common with each line of class $V$.

## Definition - Axes, Coordinates, Abscissa, Ordinate

> Given a **grid system**, any two intersecting lines $L_1$ and $L_2$ may be chosen to represent the system **axes** and referred to as the **X axis** and the **Y axis** respectively.  In the same way that a single number $x$ represents a **coordinate** or point along a line in one dimension, two numbers written as $(x, y)$ represent a coordinate in two dimensions.  $x$ is called the **abscissa** and locates a point's position along the X axis.  Similarly, $y$ is called the **ordinate** and specifies an offset from $0$ along the Y axis.  The point with coordinates $(0, 0)$ (i.e., where $L_1$ and $L_2$ intersect) is called the **origin** of the system.  

## Definition - Point Equality (2D)

> Two points $A = (a, b)$ and $B = (c, d)$, are **equal** if and only if $a=c$ and $b=d$.

## Definition - Vertical Line

> A **vertical line** is a set of points $\bigl\{ (x,y) \bigm| x\,\hbox{is a constant} \bigr\}$ or equivalently $\bigl\{(x,y)\bigm|x=h\bigr\}$ where $h$ is a constant.

## Definition - Horizontal Line

> A **horizontal line** is a set of points $\bigl\{ (x,y) \bigm| y\,\hbox{is a constant} \bigr\}$ or equivalently $\bigl\{(x,y)\bigm|y=g\bigr\}$ where $g$ is a constant.

## Definition - Distance Between Points

> The **distance** from $A = (x_1, y_1)$ to $B = (x_2, y_2)$ is
>
> $$AB = \sqrt{ (x_2 - x_1)^2 + (y_2 - y_1)^2 }.$$

Note: The textbook also uses the term *measure* to refer to a shortest path between two points or to the measure of an angle, a sometimes confusing abstraction of terminology.

### Theorem 5-1

> If $d$ is a non-negative real number and $(a,b)$ is a specific point, then there is at least one point $(m,n)$ such that $d=\sqrt{(m-a)^2+(n-b)^2}$.

### Theorem 5-2

> If two points are on a horizontal line, then the distance between them is the absolute value of the difference of their corresponding $x$ coordinates.
