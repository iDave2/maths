# Chapter 10 - Angles

This chapter introduces terminology that allows one to have precise conversations about angles and triangles that no-one else will understand.

## Definition - Angle, Vertex, Sides

> An **angle** is the union of two rays with a common endpoint.  This common endpoint is called the **vertex**, and the two rays are called the **sides** of the angle.

## Definition - Polygon

> A **polygon** is a set of $n$ line segments such that:

1. The $n$ line segments intersect in exactly $n$ points;
2. The line segments intersect only at their endpoints.

## Definition - Polygon, Vertex, Sides

> Each line segment of a polygon is called a **side** of the polygon and each point of intersection is called a **vertex**.

## Definition - Triangle

> A **triangle** is a polygon of three sides.  The three sides (in elided picture) are $\Segment{AB}$, $\Segment{AC}$, and $\Segment{BC}$, and the three vertices are $A$, $B$, and $C$.  The triangle is denoted by $\triangle ABC$.  $\angle ABC$, $\angle ACB$, and $\angle BAC$ are called the **angles** of $\triangle ABC$.

## Definition - Measure of an Angle

> The **measure** of an angle is the amount of rotation used to dscribe the angle's "opening."  The measure is usually expressed (in this book) in **degrees**.  The symbol for degrees is '$\Degrees$' and the measure of $\angle ABC$ is denoted "$m\angle ABC$".  For example, if $\angle ABC$ has a measure of $30\Degrees$, it is written $m\angle ABC = 30\Degrees$.  The measure of all angles will be a number $r$ such that
>
> $$ 0\Degrees \le r \le 180\Degrees. $$

Note this convention of the textbook and keep it in quadrants I and II.

## Definition - Right Angle

> Let two rays $\Ray{AB}$ and $\Ray{AC}$ form $\angle BAC$.  $\angle BAC$ is called a **right angle** if and only if $\Ray{AB} \perp \Ray{AC}$.  The right angle is often shown in drawings by placing a small "box" at the right angle (as shown in the elided picture).

## Postulate 6 - 90 Degrees

> An angle is a right angle if and only if it has a measure of $90\Degrees$.

Is this a postulate or a definition?

## Definition - Congruent Angles

> Two angles are **congruent** if and only if they have the same measure.

## Definition - Zero, Acute, Obtuse Angle

> Let the measure of an angle be $r$.
>
> * If $r=0$, the angle is called a **zero angle** and its two sides conincide.
> * If $0<r<90$, the angle is called an **acute** angle.
> * If $90 < r < 180$, the angle is called an **obtuse** angle.

## Definition - Interior of Angle

> Given any angle $\angle ABC$, a point $D$ lies in the **interior** of the angle if and only if some line through $D$ intersects the sides of the angle in points $X$ and $Y$ (different from the vertex) such that $\Between X D Y$; that is, $D$ is between $X$ and $Y$.

## Definition - Exterior of Angle

> A point is in the **exterior** of an angle if and only if it is:
>
> 1. not in the interior of the angle, and
> 2. not on the angle itself.

## Definition - Adjacent Angles

> Two angles are **adjacent** if and only if
>
> 1. They have the same vertex,
> 2. They have exactly one common side, and
> 3. The intersection of their interiors is the empty set.

## Definition - Straight Angle

> If two adjacent angles have their non-common sides lying on a line, those two sides form a **straight angle**.

## Definition - Interior of Triangle

> Given any triangle $\triangle ABC$, a point $D$ is in the **interior of the triangle** if and only if it is in the interior of two angles of the triangle.

## Definition - Exterior of Triangle

> A point is in the **exterior of a triangle** if and only if it is
>
> 1. not in the interior of the triangle, and
> 2. not on the triangle itself.

## Definition - Exterior Angle of Triangle

> An **exterior angle** of a triangle is an angle which is adjacent to an angle of the triangle and with it forms a straight angle.

## Postulate 7 - Angle Addition

> **Angle Addition Postulate**:  If $D$ is a point in the interior of $\angle ABC$, then $m\angle ABD + m\angle DBC = m\angle ABC$.

### Theorem 10-1

> Every angle is congruent to itself.  (Identity.)

### Theorem 10-2

> The measure of a straight angle is $180\Degrees$.

### Theorem 10-3

> The sum of the measures of the adjacent angles with a common vertex is $360\Degrees$ if:
>
> 1. each angle is adjacent to exactly two other angles and
> 2. the intersection of the interiors of any two angles is the empty set.

### Theorem 10-4

> All right angles are congruent.

### Theorem 10-5

> The adjacent angles formed by two intersecting lines are congruent if and only if the lines are perpendicular.

## Definition - Supplementary Angles

> Two angles are **supplementary** if and only if the sum of their measures is $180\Degrees$.

## Definition - Complementary Angles

> Two angles are **complementary** if and only if the sum of their measures is $90\Degrees$.

### Theorem 10-6

> Two adjacent angles whose non-common sides lie on a line are supplementary.

### Theorem 10-7

> Supplements of congruent angles are congruent.

### Theorem 10-8

> Complements of congruent angles are congruent.

### Theorem 10-9

> If two angles are congruent and supplementary, then they are right angles.

## Definition - Vertical Angles

> Two distinct intersecting lines form two pairs of non-adjacent angles.  Such pairs are called **vertical angles**.

### Theorem 10-10

> **Vertical Angle Theorem**:  Vertical angles are congruent.
