# Chapter 8 - Parallel and Perpendicular Lines

This chapter begins with a definition of parallel lines that causes pause since parallel lines were already defined [in Chapter 5](Chapter05_CoordinateSystems.ipynb#Postulate-3) as lines with no intersection.  A Wikipedia version of that definition is repeated here for reference:

## Definition - Parallel Lines [dek]

> **Parallel** lines are lines in a plane which do not meet; that is, two lines in a plane that do not intersect or touch each other at any point are said to be parallel.

Aside from breaking orthogonality rules (i.e., axioms and definitions should stand independently), multiple definitions of a single concept require proof of equivalence but this is not done.  Perhaps authors felt this theorem was too difficult or simply too long for a high school scope.

To remedy these concerns, the second definition of parallel lines is presented here as a theorem rather than a definition.

### Theorem 8-0 - Slope of Parallel Lines [dek]

> Two distinct lines in a plane are *parallel* if and only if,
>
> 1. Their slopes are equal, or
> 2. They both have no slope (i.e., they both are vertical).
>
> One writes $L_1 \Vert L_2$ when $L_1$ and $L_2$ are parallel.

#### Proof

> **Case 1**.  Suppose lines $L_1$ and $L_2$ are vertical lines; that is, they both have no slope.  Then they may be expressed as loci,
>
> $$L_1 = \bigl\{(x,y)\bigm|x=x_1\bigr\}$$
>
> where $x_1$ is constant, and
>
> $$L_2 = \bigl\{(x,y)\bigm|x=x_2\bigr\}$$
>
> where $x_2$ is constant.  Since $L_1$ and $L_2$ are distinct, $x_1 \ne x_2$ and the lines have no points in common.  Therefore, the lines are parallel.
>
> **Case 2**.  Let $L_1 = \bigl\{(x,y)\bigm|y=mx+b_1\bigr\}$, $L_2 = \bigl\{(x,y)\bigm|y=mx+b_2\bigr\}$, where $b_1\ne b_2$.  Then $L_1$ and $L_2$ have the same slope $m$.  If $L_1$ and $L_2$ share a common point $(u,v)$, then $(u,v)$ satifies both $v = mu + b_1$ and $v = mu + b_2$ which, when substracted, shows $b_1 = b_2$.  This contradiction implies $L_1$ and $L_2$ share no common points and are thus parallel.
>
> **Case 3**.  Suppose lines $L_1$ and $L_2$ are parallel, $L_1\Vert L_2$.  Then they are distinct by definition.  We show they have the same or no slope.
>
> Choose distinct points $(x_{11}, y_{11})$ and $(x_{12}, y_{12})$ on $L_1$, 
$(x_{21}, y_{21})$ and $(x_{22}, y_{22})$ on $L_2$.  There are two possibilities:
>
> 1. $x_{11} = x_{12}$. Then $L_1$ is vertical and has no slope.  Since $L_1\Vert L_2$, $L_2$ must also be vertical for otherwise it would have slope $m = (y_{22}-y_{21})/(x_{22}-x_{21})$ and intersect $L_1$ at point $(x_{11}, y)$ where $y - y_{21} = m(x_{11} - x_{21}).$  It follows that $L_2$, like $L_1$, has no slope.
>
> 2. $x_{11}\ne x_{12}$. Now $L_1$ has slope $m_1$.  If $L_2$ were vertical, it would intersect $L_1$ at its x-intercept, so $L_2$ has slope $m_2$ and we can write down equations for these lines (unfortunately, this MathJax cannot format `\eqno(1)` ...):
>
> $$ L_1 = \bigl\{ (x,y) \bigm| y = m_1 x + b_1 \bigr\}, \qquad  
     L_2 = \bigl\{ (x,y) \bigm| y = m_2 x + b_2 \bigr\}. \qquad\qquad(1) $$  
>
> (... and MarkDown loses the list after display mode.)  The difference between these equations is the set $X$ of all $x$ for which $L_1$ and $L_2$ have a point in common,
>
> $$ X = \bigl\{ x \bigm| b_1 - b_2 = (m_2 - m_1) x \bigr\}. \qquad\qquad(2) $$,
>
> and $X$ must be empty since $L_1$ and $L_2$ do not touch.  $b_1 \ne b_2$ for the same reason.  The only way for $X$ to be empty is for its condition of inclusion, its equation, to have no solution, and the only way for that to happen is if $(m_2 - m_1) = 0$.  It follows that $L_1$ and $L_2$ have the same slope and the proof is complete.

### Theorem 8-1 - Parallel Horizontal Lines

> All distinct horizontal lines are parallel.

### Theorem 8-2 - Parallel Vertical Lines

> All distinct vertical lines are parallel.

## Definition - "Exactly One"

> The phrase "exactly one" means that the two properties of *existence*
and *uniqueness* are present:
>
> 1. existence - "there is at least one"
> 2. uniqueness - "there is at most one"
>
> The phrase "one and only one" is equivalent to the phrase "exactly one."

### Theorem 8-3 - Restatement of Parallel Postulate

> Through any point P not on a given line $L_1$, there is exactly one line $L_2$ parallel to the given line $L_1$.  (Euclidean parallel postulate.)

Not to dampen the love, but the authors claimed in Chapter 2 that Euclid tried and failed to prove this using his first four axioms, yet now at long last they finally have a proof.  Their proof hinges on the concept of slope with its formalization in analytical geometry centuries after Euclid.  A more accurate presentation might say this cannot be proved synthetically, or organically, like organic vegetables.

### Theorem 8-4 - Intersecting Parallel Lines

> If a line intersects one of two parallel lines in exactly one point, then it will intersect the other line in exactly one point.

### Theorem 8-5 - Transitivity of Parallel Operator

> if lines $L_1$, $L_2$, and $L_3$ are three distinct lines such that $L_1\Vert L_2$ and
$L_2\Vert L_3$, then $L_1\Vert L_3$.

These little theorems seem primarily for homework examples.  The chapter next moves on to perpendicular lines.

## Definition - Perpendicular Lines

> Two lines are **perpendicular** if and only if,
>
> 1. The product of their slopes is -1, or
> 2. One line has a slope of zero and the other has no slope (i.e., one horizontal, the other vertical).
>
> The expression $L_1\perp L_2$ is used to indicate $L_1$ and $L_2$ are perpendicular.

### Theorem 8-6 - Perpendicular Existence and Uniqueness

> If $P$ is a point on line $L_1$, then there is exactly one line $L_2$ through $P$ such that $L_1\perp L_2$.

### Theorem 8-7 - Relationship Between Perpendicular and Parallel Lines

> If two lines are perpendicular to one line at different points, then the two lines are parallel.

### Theorem 8-8 - Relationship Between Parallel and Perpendicular Lines

> In a plane, if a line is perpendicular to one of two parallel lines, then it is also perpendicular to the other parallel line.

## Definition - Distance from Point to Line

> The **distance** from a point $P$ to a line $L_1$ is found by computing the distance along the line $L_2$ perpendicular to $L_1$ from $P$ to the point of intersection, $Q$.

#### Commentary

Distance between two points was defined in [Chapter 5](Chapter05_CoordinateSystems.ipynb#Postulate-5).  This is the first definition of the distance between a line and a point.  Its tacit assumption that $L_2$ *exists* may leave readers unsatisfied and this author would prefer the following presentation:

1. Postulate the *existence* of $L_2$. From Euclid's perspective, this simply captures his "common notion" that one may construct $L_2$ with a compass.
2. Use Euclid's [Axiom 1](Chapter02_HistoricalGeometry.ipynb#Axiom-1---Two-Points-and-One-Line) or its restatement in [Postulate 2](Chapter05_CoordinateSystems.ipynb#Postulate-2---Restatement-of-Axiom-1) to prove *uniqueness* of $L_2$.
3. Now define distance from point to line as simply the measure of segment $\Segment{PQ}$.

Since this result becomes a no-brainer this late in book, the presentation might better be placed in Chapter 5 near initial definition of point-to-point distance.  Euclid clearly had the notion of "right angle;" an initial definition of "perpendicular" in terms of his Axioms 4 and 5, prior to slopes and analysis, seems preferable.

### Theorem 8-9 - Distance Between Parallel Lines

> The distance from any point of one of two parallel lines to the other line is a constant.

The proof of all these theorems is the same; namely, solving the linear equation for a line, $Ax + By + C = 0.$
